```python
import pandas as pd # Importing the pandas library
import matplotlib.pyplot as plt # Import matplotlib.pyplot
import seaborn as sns # Import seaborn
# 1. Load the data
df = pd.read_csv("/content/Multiple disk clutch brake.csv")

# 2. Handle Missing Values (Forward Fill)
df = df.fillna(method='ffill')
df = df.fillna(0)

# 3. Normalize the Data
# Assuming you want to normalize across columns (algorithms)
# Get numeric columns only for normalization
numeric_columns = df.select_dtypes(include=['number']).columns

for column in numeric_columns:  # loop through numeric columns
    # Calculate min and max for each column
    min_val = df[column].min()
    max_val = df[column].max()

    # Apply normalization formula: (value - min) / (max - min)
    df[column] = (df[column] - min_val) / (max_val - min_val)

# Violin Plot
plt.figure(figsize=(14, 8))
sns.violinplot(data=df)
plt.title("Algorithm Objective Value Distribution (Normalized)", fontsize=16)
plt.xlabel("Multiple disk clutch brake", fontsize=14)  # Changed X-axis label
plt.ylabel("Fitness", fontsize=14)  # Changed Y-axis label
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Line Plot
plt.figure(figsize=(14, 8))
sns.lineplot(data=df)
plt.title("Algorithm Objective Value Distribution (Normalized)", fontsize=16)
plt.xlabel("Multiple disk clutch brake", fontsize=14)  # Changed X-axis label
plt.ylabel("Fitness", fontsize=14)  # Changed Y-axis label
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import os


# Get a list of all CSV files in the directory
csv_files = [f for f in os.listdir("/content/") if f.endswith(".csv")]

# Create a figure and axes for the subplots
fig, axes = plt.subplots(3, 2, figsize=(14, 12))  # 3 rows, 2 columns
axes = axes.flatten()  # Flatten the axes array for easier iteration

# Loop through each CSV file and create a violin plot
for i, file in enumerate(csv_files):
    # Load the data
    df = pd.read_csv(os.path.join("/content/", file))

    # Handle Missing Values (Forward Fill)
    df = df.fillna(method='ffill')
    df = df.fillna(0)

    # Normalize the Data
    numeric_columns = df.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        min_val = df[column].min()
        max_val = df[column].max()
        if max_val - min_val != 0:
            df[column] = (df[column] - min_val) / (max_val - min_val)
        else:
            df[column] = 0
        df[column] = df[column].replace([np.inf, -np.inf], 0)

    # Create the violin plot on the current subplot
    sns.violinplot(data=df, ax=axes[i])

    # Remove .csv from the title:
    axes[i].set_title(f"{file[:-4]}", fontsize=12)  # file[:-4] removes the last 4 characters (.csv)

    axes[i].set_xlabel("Algorithm", fontsize=10)
    axes[i].set_ylabel(" Fitness", fontsize=10)
    axes[i].tick_params(axis='x', rotation=45, labelsize=8)
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```
